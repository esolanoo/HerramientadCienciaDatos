```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy.optimize import fsolve

f = lambda x: np.exp(-x) * np.cos(8*x)
x = np.linspace(0, 4.8, 1000)
y = f(x)

fig, ax = plt.subplots(figsize=(10, 5), facecolor='#FFE5E5')
ax.set_facecolor('#E5E8CE')
ax.plot(x, y)
ax.plot(x, np.exp(-x), color='#F2B367', label=r"$e^{-x}, -e^{-x}$", linewidth=3, alpha=.8)
ax.plot(x, -np.exp(-x), color='#F2B367', linewidth=3, alpha=.8)
ax.set_title(r"f(x) = $e^{-x}$cos(8x)")
ax.set_ylabel('f(x)', rotation=0, size=12)
ax.yaxis.set_label_coords(0, 1)
ax.set_xlabel('x', size=12)
ax.xaxis.set_label_coords(1, .5)

plt.yticks([-1, 1])

ax.xaxis.set_major_locator(ticker.MultipleLocator(base=np.pi/8))

def format_func(value, tick_number):
  N = int(np.round(8 * value / np.pi))
  if N == 0:
    return ""
  elif N == 1:
    return r"$\frac{\pi}{8}$"
  elif N == 2:
    return r"$\frac{\pi}{4}$"
  elif N == 4:
    return r"$\frac{\pi}{2}$"
  elif N == 8:
    return r"$\pi$"
  elif N % 4 == 0:
    return r"$\frac{{{}\pi}}{{2}}$".format(N//4)
  elif N % 2 == 0:
    return r"$\frac{{{}\pi}}{{4}}$".format(N//2)
  else:
    return r"$\frac{{{}\pi}}{{8}}$".format(N)

ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))

ax.tick_params(axis='x', labelsize=14)

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

ax.grid(linestyle='--')
ax.yaxis.grid(False)
ax.xaxis.grid(True)
ax.set_axisbelow(False)

ax.spines[['right', 'top']].set_visible(False)

x_inter = fsolve(lambda x: abs(f(x)) - np.exp(-x), x)
y_inter = f(x_inter)

plt.plot(x_inter, y_inter, '*', markersize=3, color='#F100D9', alpha=0.5)

ax.legend()
fig.show()
```
