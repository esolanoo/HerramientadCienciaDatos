*Usar '.' como separador en las columnas está mal, lo voy a cambiar a '_'*

```python
names = [x.replace('.', '_') for x in df.columns]
df.columns = names
df
```

### Por cada teatro

- Calcular la capacidad máxima del teatro

```python
unique_theatres = df['Show_Theatre'].unique()
d = {}
for teatro in unique_theatres:
    filtered = df[df['Show_Theatre']==teatro]
    d[teatro] = filtered['Statistics_Attendance'].max()
d
```

### Por todos los teatros

- ¿Cuál es el teatro con el nombre más largo?

```python
longest = max(unique_theatres, key=lambda x: len(x))
print(f"El teatro con el nombre más largo es '{longest}'")
```

* ¿Cuál es la que tiene más palabras?

```python
longest = max(unique_theatres, key=lambda x: len(x.split()))
print(f"El teatro con más palabras es '{longest}'")
```
