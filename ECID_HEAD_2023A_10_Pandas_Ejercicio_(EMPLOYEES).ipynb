```python
import os
import pandas as pd

if 'dataset.csv' in os.listdir(os.getcwd()):
    df = pd.read_csv('dataset.csv')
    df.columns = ['index'] + list(df.columns.values)[1::]
else: 
    df = pd.DataFrame()
    folders = ['20_1', '20_2', '20_3']
    a = 0
    for folder in folders:
        for file in os.listdir(folder):
            try:
                aux = pd.read_xml(os.path.join(folder, file))
                df = pd.concat([df, aux])
            except:
                continue
    df = df.reset_index()
    df.to_csv('dataset.csv')

df.sample(5)
```

```python
departments = pd.read_xml("departments.xml")
departments
```

-  Top 50 de salarios con el departamento al que pertenecen (nombre y apellido, salario y nombre de departamento)

```python
top50 = df.nlargest(50, 'salary')
top50 = pd.merge(top50, departments, on='deptno')[['firstname', 'lastname', 'salary', 'deptname']]
top50.head(5)
```

- Cantidad de Personas por Departamento (qué departamento son y cuántas personas hay)

```python
emp_dpto = pd.merge(df, departments, on='deptno')[['empno', 'deptname']]
emp_dpto.groupby('deptname').count()
```

- ¿Cuáles son los empleados contratados más jóvenes?

```python
young = df.sort_values('birthdate', ascending=False)
young.head(5)
```

- Promedio de salario por departamento

```python
sal = pd.merge(df, departments, on='deptno')[['salary', 'deptname']]
sal.groupby('deptname').mean()
```

- Top 5 de salario en cada departamento

```python
pd.DataFrame(sal.groupby('deptname').salary.nlargest(5))
```

- Para cada departamento, dar el nombre del departamento y el nombre de su manager

```python
pd.merge(df, departments, left_on='empno', right_on='mgrno')[['deptname', 'firstname', 'lastname']].set_index('deptname')
```

- En base a los anteriores, propón alguna estadística que pudiera ser interesante.

*¿Cuáles son los 10 empleados NO gerentes que ganan más dinero? (numero de empleado, nombre, apellido, salario, y nombre de departamento)*

```python
no_mgr = pd.merge(df, departments, how='left', left_on='empno', right_on='mgrno')[['empno', 'firstname', 'lastname', 'salary', 'deptno_x']]
no_mgr = pd.merge(no_mgr, departments, how='left', left_on='deptno_x', right_on='deptno')[['empno', 'firstname', 'lastname', 'salary', 'deptname']]
no_mgr.nlargest(10, 'salary').reset_index()
```
